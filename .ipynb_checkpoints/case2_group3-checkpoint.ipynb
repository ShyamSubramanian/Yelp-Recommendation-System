## Business Objects

Business objects contain basic information about local businesses. The fields are as follows:

```json
{
  'type': 'business',
  'business_id': (a unique identifier for this business),
  'name': (the full business name),
  'neighborhoods': (a list of neighborhood names, might be empty),
  'full_address': (localized address),
  'city': (city),
  'state': (state),
  'latitude': (latitude),
  'longitude': (longitude),
  'stars': (star rating, rounded to half-stars),
  'review_count': (review count),
  'photo_url': (photo url),
  'categories': [(localized category names)]
  'open': (is the business still open for business?),
  'schools': (nearby universities),
  'url': (yelp url)
}
```
## Checkin Objects
```json
{
    'type': 'checkin',
    'business_id': (encrypted business id),
    'checkin_info': {
        '0-0': (number of checkins from 00:00 to 01:00 on all Sundays),
        '1-0': (number of checkins from 01:00 to 02:00 on all Sundays),
        ...
        '14-4': (number of checkins from 14:00 to 15:00 on all Thursdays),
        ...
        '23-6': (number of checkins from 23:00 to 00:00 on all Saturdays)
    }, # if there was no checkin for a hour-day block it will not be in the dict
}
```

# Data Collection/Processing: 

````python
#----------------------------------------------
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

def findTopBiz(theCity, business_data, checkin_counts):
    bizList = []
    for bid in business_data:
        city = business_data[bid]['city']
        if city ==theCity:
            bizList.append(bid)

    bizCounts = {}
    for biz in bizList:
        if biz in checkin_counts:
            bizCounts[biz] = checkin_counts[biz]
        else:
            bizCounts[biz]=0
    return sorted(bizCounts, reverse=True, key = lambda x: bizCounts[x])

def getTopData(theCity, business_data, checkin_counts):
    res = []
    bizList = findTopBiz(theCity, business_data, checkin_counts)
    for bid in bizList:
        res.append(business_data[bid])
    return res
````

````python
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary
````

````python
import json
````

````python
business_data = {}
with open('yelp_academic_dataset_business.json',encoding="utf8") as f:
    for line in f:
        temp = json.loads(line)
        id = temp['business_id']
        business_data[id] = temp
````

````python
# key: categoryName, value: name
categoryCounts = {}

for bid in business_data:
    business = business_data[bid]
    categories =business['categories']# .split(', ')
    if categories:
        categoryList = categories.split(', ')
        for c in categoryList:
            if c in categoryCounts:
                categoryCounts[c] += 1
            else:
                categoryCounts[c] = 1
````

````python
top_cat = sorted(categoryCounts, reverse=True, key=lambda x: categoryCounts[x])[:10]
````

````python
for c in top_cat:
    print(c, categoryCounts[c])
````

** (2) Find the most popular business objects** 
* print the top 10 most popular business objects in the dataset and their counts (i.e., how many checkins in total for each business object).  Here we say a business object is "popular" if the business object attracts a large number of checkins from the users.

````python
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary
````

````python
import json
````

````python
checkin_data = {}
with open('yelp_academic_dataset_checkin.json',encoding="utf8") as f:
    for line in f:
        temp = json.loads(line)
        id = temp['business_id']
        checkin_data[id] = temp
````

````python
#key: id, value: counts
checkin_counts = {}
for b_id in checkin_data:
    time = checkin_data[b_id]['time']
    counts = sum(time.values())
#     print(counts)
    checkin_counts[b_id] = counts
````

````python
top10 = sorted(checkin_counts, reverse=True, key = lambda x: checkin_counts[x])[:10]
for b_id in top10:
    print((b_id, checkin_counts[b_id]))
````

````python
Toronto_stars=[]
for i in range(50):
    Toronto_stars.append(Toronto_data[i]['stars'])
print(Toronto_stars)

tlen = pd.Series(data = Toronto_stars)
tlen.plot(figsize = (20,5), color = 'r');












````

````python
food_type=['Coffee & Tea','Fast Food','American (Traditional)','Italian','Mexican','Chinese','American (New)','Japanese']
review_length=[]
for bid in business_data:
    temp = business_data[bid]
    categories = temp['categories']
    length=temp['review_count']
    if categories:
        categoryList = categories.split(', ')
        for c in categoryList:
            if c in food_type:
                review_length.append(length)
length=len(review_length)
review_length=sorted(review_length, reverse=True)
a1=review_length[int(length/5)]
a2=review_length[2*int(length/5)]
a3=review_length[3*int(length/5)]
a4=review_length[4*int(length/5)]

````

````python
tlen = pd.Series(data = review_length)
tlen.plot(figsize = (20,5), color = 'r');
````

````python
food_type=['Coffee & Tea','Fast Food','American (Traditional)','Italian','Mexican','Chinese','American (New)','Japanese']
food_cate=[]
stars=[]
#data = pd.DataFrame(data = [text for text in text], columns = ['Tweet_BJ'])
for bid in business_data:
    temp = business_data[bid]
    categories = temp['categories']# .split(', ')
    star=temp['stars']
    length=temp['review_count']
    if categories:
        categoryList = categories.split(', ')
        for c in categoryList:
            if c in food_type:
                food_cate.append(c)
                if length<=a1:   
                    stars.append(star*0.6)
                elif a1<length<=a2:   
                    stars.append(star*0.7)
                elif a2<length<=a3:   
                    stars.append(star*0.8)
                elif a3<length<=a4:   
                    stars.append(star*0.9)
                elif a4<length: 
                    stars.append(star*1)
````

````python
data = pd.DataFrame(data = [c for c in food_cate], columns = ['food_type'])
data['stars']  = np.array([c for c in stars])
sns.set(style='ticks')
f, ax = plt.subplots(figsize=(10, 10))
sns.boxplot(x ='stars', y = 'food_type', data = data, whis='range', palette='vlag')
ax.set(ylabel='')
ax.xaxis.grid(True)
sns.despine(trim=True, left=True)
````
